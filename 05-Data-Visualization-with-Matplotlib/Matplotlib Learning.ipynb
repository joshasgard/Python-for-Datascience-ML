```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

#allows visualizing the plot in this notebook

```python
import numpy as np
x= np.linspace(0,5,11)
y = x ** 2
```

```python
x, y
```

```python
#Functional method of plotting
plt.plot(x,y,'r','--')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object-Oriented (OO) Matplotlib
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabel')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #[left, bottom,width,height]
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

# Matplotlib - Part 2

```python
fig,axes = plt.subplots(nrows =1, ncols =2)
#plt.tight_layout() #removes overlapping in plots

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1,figsize=(8,2)) 

#(width, height)in inches

axes[0].plot(x,y)

axes[1].plot(y,x)
#dpi is dots per inch
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, color = 'Purple',label='X squared', linewidth = 10, 
        alpha = 0.5)
ax.plot(x,x**3,color = '#FF8C00', label = 'X cubed',linestyle='-.', 
        marker = '.', markersize = 20, markerfacecolor='green',markeredgecolor='red')
#linewidth is also lw, linestyle is ls
ax.legend(loc = (0.1,0.8))
ax.set_title('Title')
```
